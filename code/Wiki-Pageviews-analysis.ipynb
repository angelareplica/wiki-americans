```python
# who's the top-viewed person by city?

df1 = city_name_df.groupby(['city', 'name'], as_index=False)['views'].sum()
df2 = df1.sort_values(['city','views'])
df2 = (df2.set_index(['city', 
                      df2.groupby(['city']).cumcount(ascending=False)])['name']
          .unstack()
          .add_prefix('max_'))

city_mostviewedperson_df = df2['max_0'].reset_index()

city_mostviewedperson_df
```

```python
# add the number of views to the dataframe
city_mostviewedperson_df['name'] = city_mostviewedperson_df['max_0']
```

```python
# add the number of views to the dataframe
city_mostviewedperson_df2 = pd.merge(city_mostviewedperson_df, city_name_df, how='left', on=['name', 'city'])
city_mostviewedperson_df2
```

```python
city_mostviewedperson_df2 = city_mostviewedperson_df2.drop(columns=['name'])
```

```python
# save to CSV
city_mostviewedperson_df2.to_csv("most_viewed_person_by_city.csv", index=False)
```
